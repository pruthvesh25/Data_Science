```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn.metrics import accuracy_score
```

```python
df = pd.read_csv('diabetes.csv')
```

```python
df
```

```python
df.shape
df.describe()
```

```python
df['Outcome'].value_counts()
```

```python
# 0 - non diabetic 
#  - diabetic
```

```python
x = df.drop(columns='Outcome',axis=1)
y = df['Outcome']
```

```python
x
```

```python
y
```

```python
# standerdization
```

```python
scaler = StandardScaler()
```

```python
scaler.fit(x)
```

```python
standardized_data = scaler.transform(x)
```

```python
standardized_data
```

```python
x = standardized_data
```

```python
x
```

```python
y
```

```python
x_train , x_test ,  y_train , y_test = train_test_split(x,y,test_size=0.2)
```

```python
x_train.shape
```

```python
clf = svm.SVC(kernel='linear')
```

```python
clf.fit(x_train,y_train)
```

```python
x_train_predict = clf.predict(x_train)
```

```python
accuracy_score(x_train_predict,y_train)
```

```python
# for test data 
```

```python
x_test_predict = clf.predict(x_test)
accuracy_score(x_test_predict,y_test)
```

```python
for i in range(114):
    if(x_test_predict[i]==0):
        print("Person",i+1," is not diabetic")
    else:
        print("Person ",i+1," is diabetic")
```
